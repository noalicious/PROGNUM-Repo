```python
#question 3.2

from scipy.constants import G

masses = [1.9891e+30, 1.8986e+27, 5.6846e+26, 
          1.0243e+26, 8.6810e+25, 5.9736e+24, 
          4.8685e+24, 6.4185e+23, 3.3022e+23, 
          7.349e+22, 1.25e22]
    
radii  = [696340, 71492, 60268, 24764, 
          25559, 6378.1, 6051.8, 3396.2, 
          2439.7, 1738.1, 1195]

names  = ['Sun', 'Jupiter', 'Saturn', 'Neptune', 
          'Uranus', 'Earth', 'Venus', 'Mars', 
          'Mercury', 'Moon', 'Pluto']

n = len(masses)
list = range(n)
Vesc_list=[]
for i in list:
    M = masses[i]
    r = radii[i]
    Vesc = ((2*G*M/r)**0.5)
    Vesc_list.append(Vesc)
    print(f'with {Vesc} m/s : Mass is {masses[i]:.2E}')


Dictionary = dict(zip(names, Vesc_list))
print(Dictionary)
```

```python
#Question 3.4 coursework

#the full code is calculating the standard deviation for list a without negative entries
a= [2.3, 3.5, 5.6, 12.8, 10.3, 7.2, 5.0, 2.1, -2.0]
data=[i for i in a if i>=0]    #excluding negative entries

#calculating the sum of list: Data
sum = 0
for val in data:
	sum += val

#calculating the number of entries in list: Data
N = 0
for i in data:
    N += 1
    
#calculating the average of the list data
Mu = (sum)/N

#calculating the standard deviation  
x=0
for i in data:
    x += ((i-Mu)**2)

sigma = ((1/N)*(x))**0.5

print(f'the mean of the list is: {Mu}')
print(f'the standard deviation is: {sigma}')



```

```python
#Question 3.7
import math

def angular_size(d, D):
    ratio = d / (2 * D)
    radians = 2 * math.atan(ratio)
    degrees = math.degrees(radians)
    return degrees

A = float(input('The size of the object is: '))
B = float(input('The distance to the object: '))
result = angular_size(A, B)

print(f"The angular size is: {result:} degrees")
```

```python
#Question 3.10
from math import pi

def cylinder(r, h):
    area = 2 * pi * r**2 + 2 * pi * r * h
    volume = pi * r**2 * h
    return area, volume


radius = 5
height = 10
a, v = cylinder(radius, height)
ratio = a / v

print(f"Area: {a}")
print(f"Volume: {v}")
print(f"Ratio (Area/Volume): {ratio}")
```

```python
#question 3.17
import numpy as np

def stdev(y):
    mu = np.mean(y)
    sigma = np.sqrt(np.mean((y - mu)**2))
    return sigma

data = np.array([1, 2, 3, 4, 5])

print(f"standard deviation: {stdev(data)}")

print(f"standard deviation of numpy:  {np.std(data)}")
```

```python
#Question 3.18
import numpy as np


x = np.array([18.49, 35.48, 35.48, 11.34, 7.12, 9.82, 23.66, 44.0, 66.7, 52.4, 55.0, 25.6, 26.19])
y = np.array([2094.75, 2320.87, 1224.9, 872.13, 432.2, 751.0, 1410.0, 2700.0, 5292.0, 2550.0, 5253.0, 2419.0, 2158.0])

x_mean = np.mean(x)
y_mean = np.mean(y)

numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean)**2)
b = numerator / denominator


a = y_mean - b * x_mean

print(f"(a): {a}")
print(f"(b): {b} km/s/Mpc")
```
