```python
#Question 2.12, 
#atan2(y, x): it accepts two arguments: x and y, it goes from -180 to 180
#atan only accepts 1 argument (y/x), it goes only from 90 to -90
from math import *
x = float(input("x-value: "))
y = float(input("y-value: "))
          
angle = degrees(atan2(y, x))
print(f"The angle is: {angle} degrees")

if 0 < angle < 90:
    print("The quadrant of (x,y) is 1")
elif 90 < angle < 180:
    print("The quadrant of (x,y) is 2")
elif -180 < angle < -90:
    print("The quadrant of (x,y) is 3")
elif -90 < angle < 0:
    print("The quadrant of (x,y) is 4")
```

```python

night1 = [1, 3, 2, 4, 0, 3, 7, 5, 2, 9, 2, 4, 3]
night2 = [0, 0, 2, 4, 2, 4, 3, 2, 1, 1, 1, 2, 0]


set1 = set(night1)       # Convert lists to sets 
set2 = set(night2)       # Convert lists to sets 

intersection_list = list(set1.intersection(set2)) # Values present in BOTH periods


difference_list = list(set1.difference(set2)) # Values in night1 that are NOT in night2


union_list = list(set1.union(set2)) #  All unique values from both periods combined

# Results
print(f"Intersection: {intersection_list}")
print(f"Difference: {difference_list}")
print(f"Union: {union_list}")
```
